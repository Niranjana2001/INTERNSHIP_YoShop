#### 5. Python program for Extract mobile no from PDF, Json,XML and MS word file and save into MS excel 

```python
import json
import os
import re

root_folder=r'C:\Users\niran\OneDrive\Desktop\Internship\contact data (1)'
for folder_path, _,filenames in os.walk(root_folder):
    for filename in filenames:
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            
            # Read the JSON file
            with open(file_path) as file:
                json_data = json.load(file)
                
            # Extract mobile number using regular expression
            pattern = r'\b\d{10}\b'
            for message in json_data['messages']:
                if 'msg' in message and "+91-9080749858" not in message['msg']:
                    match = re.search(pattern, message['msg'])
                    if match:
                        mobile_number = match.group()
                        mobile_numbers.append(mobile_number)
df = pd.DataFrame({'Mobile Number': mobile_numbers})

df.to_excel('mobile_numbers.xlsx', index=False)
```
